```python
import tkinter as tk

# Class
class DeliveryService:
    def __init__(self, location, weight):
        self.location = location
        self.weight = weight

    def calculate_fee(self):
        if self.location.lower() == "pau":
            if self.weight >= 10:
                return 2000
            else:
                return 1500

        elif self.location.lower() == "epe":
            if self.weight >= 10:
                return 5000
            else:
                return 4000

        else:
            return None


# Function for GUI
def calculate():
    location = entry_location.get()
    
    try:
        weight = float(entry_weight.get())
    except ValueError:
        output_label.config(text="Please enter a valid weight.")
        return

    service = DeliveryService(location, weight)
    fee = service.calculate_fee()

    if fee is None:
        output_label.config(text="Invalid location! Use PAU or Epe.")
    else:
        output_label.config(text=f"Delivery Fee: ₦{fee}")


# GUI Setup
root = tk.Tk()
root.title("Delivery Fee Calculator")
root.geometry("400x300")

# Title
title = tk.Label(root, text="Delivery Service System", font=("Arial", 14))
title.pack(pady=10)

# Location Input
label_location = tk.Label(root, text="Enter Location (PAU or Epe):")
label_location.pack()
entry_location = tk.Entry(root)
entry_location.pack(pady=5)

# Weight Input
label_weight = tk.Label(root, text="Enter Package Weight (kg):")
label_weight.pack()
entry_weight = tk.Entry(root)
entry_weight.pack(pady=5)

# Button
btn = tk.Button(root, text="Calculate Fee", command=calculate)
btn.pack(pady=15)

# Output
output_label = tk.Label(root, text="")
output_label.pack(pady=10)

# Run app
root.mainloop()
```
