```python
import tkinter as tk
import random

# Employee Class
class Employee:
    def __init__(self, name):
        self.name = name

        self.employees = [
            "Mary Evans", "Eyo Ishan", "Durojaiye Dare", "Adams Ali",
            "Andrew Ugwu", "Stella Mankinde", "Jane Akibo", "Ago James",
            "Michell Taiwo", "Abraham Jones", "Nicole Anide", "Kosi Korso",
            "Adele Martins", "Emmanuel Ojo", "Ajayi Fatima"
        ]

        self.tasks = [
            "Loading", "Transporting", "Reviewing Orders",
            "Customer Service", "Delivering Items"
        ]

    # Check if employee exists
    def check_employee(self):
        return self.name in self.employees

    # Take attendance
    def take_attendance(self):
        return f"Attendance taken for {self.name}"

    # Assign random task
    def assign_task(self):
        task = random.choice(self.tasks)
        return f"Task assigned: {task}"

    # Refuse access
    def refuse_access(self):
        return "Access Denied! You are not an employee."


# GUI Function
def process_employee():
    name = entry.get()

    emp = Employee(name)

    if emp.check_employee():
        result = emp.take_attendance() + "\n" + emp.assign_task()
    else:
        result = emp.refuse_access()

    output_label.config(text=result)


# GUI Setup
root = tk.Tk()
root.title("Employee Verification System")
root.geometry("400x300")

# Title
title = tk.Label(root, text="Employee Attendance System", font=("Arial", 14))
title.pack(pady=10)

# Entry
entry = tk.Entry(root, width=30)
entry.pack(pady=10)

# Button
btn = tk.Button(root, text="Verify Employee", command=process_employee)
btn.pack(pady=10)

# Output Label
output_label = tk.Label(root, text="", wraplength=300, justify="center")
output_label.pack(pady=20)

# Run app
root.mainloop()
```
